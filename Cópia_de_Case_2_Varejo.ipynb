```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] =  vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas["estado"].fillna("MS",inplace=True)
```

```python
vendas.info()
```

```python
media_preco=vendas["Preo"].mean()
vendas["Preo"].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas_correto["Mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby("Mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how= "left", on="cliente_Log")
vendas_cliente.head()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how= "left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda").renda.mean().sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending = False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,6))

plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="lightblue")

plt.ylabel("idade mdia")

plt.xlabel("bandeira")

plt.title("idade mdia por bandeira", loc="center",fontsize=20,fontweight= "bold",color= "black");
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda["idcanalvenda"],agg_idcanal_renda["renda"],color="green")
plt.ylabel("Renda mdia")
plt.xlabel("canal de venda")
plt.title("Renda mdia por acnal de venda",loc="center",fontsize=18,fontweight="bold",color="black");
```

```python
vendas_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(15,6))

plt.xlabel("data da venda")
plt.ylabel("quantidade de vendas")
x = vendas_por_data["Data"]
y = vendas_por_data["idcompra"]
plt.plot(x,y, color = "green", linewidth ="2")
plt.show()
```

```python
agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")["Preo_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(30,20))

plt.bar(agg_dept_preco["Nome_Departamento"], agg_dept_preco["Preo_com_frete"], color="lightblue")

plt.ylabel("Preo mdio")

plt.title("Preo mdio por departamento", loc="center", fontsize=18, fontweight="bold", color="black");
```
